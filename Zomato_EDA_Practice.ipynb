```python
data = pd.read_excel("Country-Code.xlsx")
data.head()
country_code_df = pd.DataFrame(data)
print(country_code_df)
```

```python
final_df = pd.merge(df, country_code_df, how='left', on="Country Code")
```

```python
country_names = final_df.Country.value_counts().index
```

```python
country_val = final_df.Country.value_counts().values
```

```python
plt.pie(country_val[:3], labels=country_names[:3], autopct='%1.2f%%')
```

```python
##Observation: Zomato's maximum transactions are from india only following by USA & UK
```

```python
final_df.columns
```

```python
ratings = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0: 'Ratings count'})
```

```python
##Observations:
# 1) When rating is in between 4.5-4.9 reflects that it is excellent
# 2) When ratings are between 4.0-4.5 reflects that it is very good
# 3) When ratings are between 3.5-3.9 reflects that it is good
# 4) When ratings are between 3.0-3.4 reflects that it is average
# 5) When ratings are between 2.5-2.9 reflects that it is average
# 6) When ratings are between 2.0-2.4 reflects that it is poor
```

```python
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (15,6)
sbn.barplot(x="Aggregate rating", y="Ratings count", data=ratings, hue='Rating color', palette=['White', 'Red', 'Orange', 'Yellow', 'Green', 'Green'])
```

```python
sbn.countplot(x='Rating color', data=ratings, palette=['blue', 'Red', 'Orange', 'Yellow', 'Green', 'Green'])
```
